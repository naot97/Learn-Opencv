##Import library and read image

```
import cv2
import numpy as np 
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
import copy 

from google.colab import files
```

```
#load image
img = cv2.imread('lena.png')
print(img.shape)
```

##Mean filter

```
# Convolution
conv = cv2.blur(img,(5,5))

stack = np.hstack([img,conv])
cv2_imshow(stack)
cv2.waitKey(0)
```

##Gaussian filter

```
# Gaussian Implement
s = 7
kernel = cv2.getGaussianKernel(s*s,0).ravel()
isize,jsize,_ = img.shape
gauss = copy.copy(img)
for i in range(isize-s):
  for j in range(jsize-s):
    for k in range(3):
      m = img[i:i+s,j:j+s,k].flatten()
      new = m * kernel
      gauss[i + int(s/2), j + int(s/2) , k] = int(new.sum())

# Gaussian by library
gauss_lib = cv2.GaussianBlur(img,(7,7),0)
stack = np.hstack([gauss,gauss_lib])
cv2_imshow(stack)
```

##Bilateral filter

```
blur = cv2.bilateralFilter(img,9,75,75)
cv2_imshow(blur)
```
