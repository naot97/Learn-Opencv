```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import copy
```

```
img = cv2.imread('messi.jpg',1)
blur = cv2.GaussianBlur(img,(5,5),0)
cv2_imshow(img)
```

## Implement Image pyramids

```
matrix = copy.copy(img)
m,n,_ = img.shape

for i in range(2):
  cv2_imshow(matrix)
  blur = cv2.GaussianBlur(matrix,(5,5),0)
  list = blur[1::2,1::2,:]
  matrix = list.astype(np.uint8)

cv2_imshow(matrix)

for i in range(2):
  m,n,_ = matrix.shape
  list = np.zeros((m*2,n*2,3))
  blur = cv2.GaussianBlur(matrix,(3,3),0).astype(np.uint8)
  list[1::2,1::2,:] = matrix
  list[0::2,0::2,:] = blur
  matrix = list.copy()
  cv2_imshow(matrix)
```
