```
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
from math import sqrt
```

```
img = cv2.imread('dave.jpg',0)
cv2_imshow(img)
```

## Implement sobel

```
m,n = img.shape
result_img = np.zeros_like(img)

for i in range(1,m-1):
  for j in range(1,n-1):
    sum = 0
    for ii in range(-1,2):
      for jj in range(-1,2):
        v = 3
        if ii == 0 : v = 0
        if ii >  0 : v = v * -1
        if ii <  0 : v = v * 1
        if jj == 0 : v = v * 3.3
        sum = sum + img[i + ii, j + jj]*v
    
    result_img[i,j] = int(sum/600)

cv2_imshow(result_img)
cv2_imshow(cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3))
```

## Sobel


```
sol_y = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
sol_x = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sol = np.sqrt(sol_x**2 + sol_y**2)
cv2_imshow(sol)
```
